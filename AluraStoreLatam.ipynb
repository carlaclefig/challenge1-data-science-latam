# 1. Análisis de facturación



```python
tiendas = {
"tienda1" : tienda1,
"tienda2" : tienda2,
"tienda3" : tienda3,
"tienda4" : tienda4
}

ingreso_total = {}

for tienda, datos in tiendas.items():
  ingreso_total[tienda] = datos["Precio"].sum()

print("Ingresos totales por tienda:")
for tienda, total in ingreso_total.items():
    print(f"{tienda}: ${total:.2f}")
```

```python
import matplotlib.pyplot as plt

nombres_tiendas = list(ingreso_total.keys())
valores_ingresos = list(ingreso_total.values())

plt.figure(figsize=(8, 5))
colores = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99'] 
plt.bar(nombres_tiendas, valores_ingresos, color=colores)

plt.xlabel("Tiendas")
plt.ylabel("Ingresos Totales")
plt.title("Ingresos por Tienda")
plt.xticks(rotation=45)
plt.tight_layout()

for i, valor in enumerate(valores_ingresos):
    plt.text(i, valor + max(valores_ingresos) * 0.01, f"${valor:,.2f}", ha='center', fontsize=9)

plt.show()
```

La tienda1 obtuvo un valor total de ingreso de $1150880400.00, lo que indica que es la tienda con más ventas.
La tienda2 y la tienda3 con un valor total de ingresos no muy altos pero tampoco tan bajos, lo que indica que aún obtiene ventas regulares.
La tienda3 obtuvo un valor total de ingreso de $1038375700.00, que es la tienda que obtiene las ventas más bajas. 

# 2. Ventas por categoría

```python
ventas_por_categoria = {}

for nombre, columna in tiendas.items():
    ventas_por_categoria[nombre] = columna['Categoría del Producto'].value_counts()
   
for tienda, conteo in ventas_por_categoria.items():
    print(f"\nCategorías más populares en {tienda}:")
    print(conteo)
```

```python
for tienda, conteo in ventas_por_categoria.items():
    top_categoria = conteo.idxmax()
    cantidad = conteo.max()
    print(f"{tienda} - Más vendida: {top_categoria} ({cantidad} ventas)")
```
