# 1. Análisis de facturación



```python
tiendas = {
"tienda1" : tienda1,
"tienda2" : tienda2,
"tienda3" : tienda3,
"tienda4" : tienda4
}

ingreso_total = {}

for tienda, datos in tiendas.items():
  ingreso_total[tienda] = datos["Precio"].sum()

print("Ingresos totales por tienda:")
for tienda, total in ingreso_total.items():
    print(f"{tienda}: ${total:.2f}")
```

```python
import matplotlib.pyplot as plt

nombres_tiendas = list(ingreso_total.keys())
valores_ingresos = list(ingreso_total.values())

plt.figure(figsize=(8, 5))
colores = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99'] 
plt.bar(nombres_tiendas, valores_ingresos, color=colores)

plt.xlabel("Tiendas")
plt.ylabel("Ingresos Totales")
plt.title("Ingresos por Tienda")
plt.xticks(rotation=45)
plt.tight_layout()

for i, valor in enumerate(valores_ingresos):
    plt.text(i, valor + max(valores_ingresos) * 0.01, f"${valor:,.2f}", ha='center', fontsize=9)

plt.show()
```

La tienda1 obtuvo un valor total de ingreso de $1150880400.00, lo que indica que es la tienda con más ventas.
La tienda2 y la tienda3 con un valor total de ingresos no muy altos pero tampoco tan bajos, lo que indica que aún obtiene ventas regulares.
La tienda3 obtuvo un valor total de ingreso de $1038375700.00, que es la tienda que obtiene las ventas más bajas. 

# 2. Ventas por categoría

```python
ventas_por_categoría = {}

for nombre, columna in tiendas.items():
    ventas_por_categoría[nombre] = columna['Categoría del Producto'].value_counts()
   
for tienda, conteo in ventas_por_categoría.items():
    print(f"\nCategorías más populares en {tienda}:")
    print(conteo)
```

```python
for tienda, conteo in ventas_por_categoría.items():
    top_categoría = conteo.idxmax()
    cantidad = conteo.max()
    print(f"{tienda} - Categoría más vendida: {top_categoría} ({cantidad} ventas)")
```

Entonces en las 4 tiendas la categoría más vendida son los Muebles, pero se puede ver que la tienda que más destaca en esta categoría es la tienda3 con 499 ventas.

# 3. Calificación promedio de la tienda


```python
promedios = {}

for nombre, df in tiendas.items():
    promedios[nombre] = df["Calificación"].mean()

for tienda, promedio in promedios.items():
    print(f"{tienda}: Calificación promedio = {promedio:.2f}")
```

```python
tiendas_para_promedio = list(promedios.keys())
calificaciones = list(promedios.values())

plt.figure(figsize=(8, 5))

plt.bar(tiendas_para_promedio, calificaciones, color=colores)

plt.plot(tiendas_para_promedio, calificaciones, marker='o', color='red', linewidth=2, label="Tendencia")

for i, calificación in enumerate(calificaciones):
    plt.text(i, calificación + 0.05, f"{calificación:.2f}",
             ha='center', fontsize=10, color='black')

plt.title("Calificación Promedio por Tienda", fontsize=14)
plt.ylim(0, 5.2)
plt.ylabel("Calificación Promedio")
plt.xlabel("Tienda")
plt.legend()
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
mas_vendidos_lista = []
menos_vendidos_lista = []

for nombre, df in tiendas.items():
    conteo = df['Producto'].value_counts()

    # Producto más vendido
    producto_mas_vendido = conteo.idxmax()
    cantidad_mas_vendido = conteo.max()

    precio_unitario_mas = df[df['Producto'] == producto_mas_vendido]['Precio'].iloc[0]
    total_mas_vendido = cantidad_mas_vendido * precio_unitario_mas

    mas_vendidos_lista.append({
        "Tienda": nombre,
        "Producto": producto_mas_vendido,
        "Cantidad Vendida": cantidad_mas_vendido,
        "Precio Total": total_mas_vendido
    })

    # Producto menos vendido
    producto_menos_vendido = conteo.idxmin()
    cantidad_menos_vendido = conteo.min()

    precio_unitario_menos = df[df['Producto'] == producto_menos_vendido]['Precio'].iloc[0]
    total_menos_vendido = cantidad_menos_vendido * precio_unitario_menos

    menos_vendidos_lista.append({
        "Tienda": nombre,
        "Producto": producto_menos_vendido,
        "Cantidad Vendida": cantidad_menos_vendido,
        "Precio Total": total_menos_vendido
    })

producto_mas_vendidos_df = pd.DataFrame(mas_vendidos_lista)
producto_menos_vendidos_df = pd.DataFrame(menos_vendidos_lista)
```

```python
print("Productos más vendidos por tienda:")
print(producto_mas_vendidos_df)
```

```python
print("Productos menos vendidos por tienda:")
print(producto_menos_vendidos_df)
```

```python
import seaborn as sns

combinado_df = pd.concat([producto_mas_vendidos_df, producto_menos_vendidos_df])

plt.figure(figsize=(10, 6))
gráfico = sns.barplot(
    data=combinado_df,
    x="Tienda",
    y="Cantidad Vendida",
    hue="Producto",     
    palette="Set2"   )      

plt.title("Productos más y menos vendidos por tienda")
plt.ylabel("Cantidad Vendida")
plt.xlabel("Tiendas")
plt.legend(title="Producto", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
promedios_lista = []

for nombre, columna in tiendas.items():
        promedio = columna['Costo de envío'].mean()
        promedios_lista.append((nombre, promedio))

promedios_envío_df = pd.DataFrame(promedios_lista, columns=["Tienda", "Costo Envío Promedio"])
print(promedios_envío_df)
```

```python
plt.figure(figsize=(9, 5))

promedios_envío_df["Color"] = promedios_envío_df["Tienda"]
gráfico = sns.barplot(
    data=promedios_envío_df.sort_values("Costo Envío Promedio", ascending=True),
    y="Tienda",
    x="Costo Envío Promedio",
    hue="Color",
    palette="Set2" ,
)

for i, row in promedios_envío_df.sort_values("Costo Envío Promedio", ascending=True).iterrows():
    plt.text(row["Costo Envío Promedio"] + 0.2, i, f"${row['Costo Envío Promedio']:.2f}", va='center', fontsize=9)

plt.title("Costo de Envío Promedio por Tienda", fontsize=14)
plt.xlabel("Costo de Envío Promedio")
plt.ylabel("Tienda")
plt.show()
```
